```python

from pathlib import Path
import os
import glob
import librosa
import pandas as pd

```

```python
wav_dir = Path('/content/drive/MyDrive/Speech Emotion Detection/Crema')
filepaths = list(wav_dir.glob(r'**/*.wav'))
```

```python
labels = list(map(lambda x: os.path.split(x)[1].split('_')[2], filepaths))
set(labels)
```

```python
filepaths = pd.Series(filepaths, name='Filepath').astype(str)
labels = pd.Series(labels, name='Label')

audio_df = pd.concat([filepaths, labels], axis=1)
audio_df
```

```python
# Initialize lists to store file names and durations
durations = []

for file in filepaths:
  audio_data, sample_rate = librosa.load(file, sr=None)
  duration = librosa.get_duration(y=audio_data, sr=sample_rate)
  durations.append(duration)

audio_df['Durations'] = durations
```

```python
# Get a list of all files in the directory
file_names = os.listdir(wav_dir)

# Filter the list to include only .wav files
wav_files = [file for file in file_names if file.endswith('.wav')]

audio_df['FileNames'] = wav_files
```

```python
audio_arrays = []

for i in audio_df['Filepath']:
    x, sr = librosa.load(i, sr=8000)
    audio_arrays.append(x)

audio_df['RawAudio'] = audio_arrays
audio_df
```
